```python
import numpy as np

# Input adjacency matrix
n = int(input("Enter number of vertices: "))
adj_matrix = np.zeros((n, n))

for i in range(n):
    print(f'Enter outgoing links for node {i}')
    outgoing_links = list(map(int, input().split()))
    for j in outgoing_links:
        adj_matrix[i][j] = 1

outgoing_links = {}
incoming_links = {}

for i in range(n):
    for j in range(n):
        if adj_matrix[i][j] == 1:
            if i not in outgoing_links:
                outgoing_links[i] = []
            outgoing_links[i].append(j)
            if j not in incoming_links:
                incoming_links[j] = []
            incoming_links[j].append(i)

# Dampening factor
d = 0.85

def page_rank_graph(d=0.85, epsilon=1.0e-6):
    # Initialize page rank
    pr = np.ones(n) # [1, 1, 1]
    delta = 1 # Initialize delta
    iteration = 0 # Initialize iteration
    print(f'Iteration {iteration}: {pr}')
    while delta > epsilon:
        pr_new = np.zeros(n)
        for i in range(n):
            sum_pr = 0
            for j in incoming_links[i]:
                sum_pr += pr[j] / len(outgoing_links[j])
            pr_new[i] = (1 - d) + d * sum_pr
        delta = np.linalg.norm(pr_new - pr)
        pr = pr_new
        iteration += 1
        print(f'Iteration {iteration}: {pr}')
    return pr, iteration

print(page_rank_graph())
```

```python
def page_rank_matrix(D, epsilon=1e-6):
	# Normalise the adjacency matrix
	A = adj_matrix
	for i in range(n):
		if np.sum(A[i]) != 0:
			A[i] = A[i] / np.sum(A[i])
	A = np.transpose(A)
	# Calculate the page rank
	pr = D
	delta = 1
	iteration = 0
	print(f'Iteration {iteration}: {pr}')
	while delta > epsilon:
		# Multiply transpose with page rank
		pr_new = np.dot(A, pr)
		delta = np.linalg.norm(pr_new - pr)
		pr = pr_new
		iteration += 1
		print(f'Iteration {iteration}: {pr}')
	return pr, iteration

# Make dampening factor vector (n * 1)
D = np.array([d for i in range(n)])
print(page_rank_matrix(D))
```

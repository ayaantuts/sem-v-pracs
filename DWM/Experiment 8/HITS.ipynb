```python
import numpy as np

# Input adjacency matrix
n = int(input("Enter number of vertices: "))
adj_matrix = np.zeros((n, n))

for i in range(n):
	adj_matrix[i] = list(map(int, input(f"Enter adjacency list for vertex {i}: ").split()))
print('Adjacency Matrix:', adj_matrix)
# Transpose of the adjacency matrix for authority calculation
adj_matrix_T = np.transpose(adj_matrix)

def normalize(vector):
	# Normalize the vector by dividing by total number of 1s in adjacency matrix
	norm = sum(vector)
	return vector / norm if norm != 0 else vector

# Initialize authority and hub scores
authority = np.ones(n)
hub = np.ones(n)
print('Initial Authority Scores:', authority)
print('Initial Hub Scores:', hub)
# Convergence threshold
epsilon = 1e-6
delta = 1
iteration = 10

while delta > epsilon:
	# Update authority score
	old_authority = authority
	old_hub = hub
	authority = [sum(hub[j] for j in range(n) if adj_matrix_T[i][j] == 1) for i in range(n)]
	authority = normalize(authority)
	# Update hub score
	hub = [sum(authority[j] for j in range(n) if adj_matrix[i][j] == 1) for i in range(n)]
	hub = normalize(hub)
	delta =	max(max(abs(authority - old_authority)), max(abs(hub - old_hub)))
	# Normalize authority scores
	iteration += 1
	print(f"Iteration {iteration}")
	print("Authority Scores:", authority)
	print("Hub Scores:", hub)
print(f"Converged in {iteration} iterations")
print("Final Authority Scores:", authority)
print("Final Hub Scores:", hub)
```
